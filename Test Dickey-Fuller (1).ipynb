```python
##ETS DECOMPOSITION
#import libs
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
# Load Specific forecasting tools
from statsmodels.tsa.arima_model import ARMA, ARMAResults, ARIMA, ARIMAResults
#from statsmodels.tsa.arima.model.ARIMA import ARMA, ARMAResults, ARIMA, ARIMAResults

from statsmodels.graphics.tsaplots import plot_acf,plot_pacf 

%matplotlib inline
tsr = pd.read_csv('/home/alessio/Scaricati/Richieste_al_minuto.csv', header=0, parse_dates=[0], index_col=0)
tsr.plot()
```

```python
#Check stationarity (p_value < 0.05)
from statsmodels.tsa.stattools import adfuller
adft = adfuller(tsr,autolag='AIC')

#print them with info
print ('Augmented Dickey-Fuller Test Result')

out = pd.Series(adft[0:4],index=['ADF test statistic','p-value','#lags used','#observations'])

for key,val in adft[4].items():
    out[f'critical value ({key})']=val
print(out)
```
